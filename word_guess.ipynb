```python
# =========================
# WORD GUESS ‚Äî minimal-change cleanup (same functions & behavior)
# =========================
import random, time
from ipywidgets import Button, Text, Label, HBox, VBox, Output, IntProgress, HTML as WHTML, Layout
from IPython.display import display, HTML, clear_output

# -------------------------
# 1) BASIC SETTINGS
# -------------------------
TITLE_TEXT, TITLE_SIZE, WORD_SIZE = "Word Guess", 72, 60
CARD_WIDTH, BUTTON_WIDTH, SUBMIT_WIDTH, ROW_GAP = 1024, 320, 220, 18
ATTEMPTS_EASY, ATTEMPTS_HARD = 5, 4
REVEAL_EASY,   REVEAL_HARD   = 1, 3

easy_words = {
   "cat":{"hint":"It's an animal","sentence":"Small house pet that says meow"},
   "bus":{"hint":"It's a vehicle","sentence":"Public transport that carries many passengers"},
   "apple":{"hint":"A fruit","sentence":"Crunchy fruit often red or green"},
   "pizza":{"hint":"Italian food","sentence":"Round, cheesy dish with many toppings"},
   "teacher":{"hint":"A profession","sentence":"Leads a class and explains lessons"},
   "doctor":{"hint":"A profession","sentence":"Treats sick people at a hospital"},
   "nurse":{"hint":"A profession","sentence":"Cares for patients and assists doctors"},
   "chef":{"hint":"A profession","sentence":"Cooks meals in a restaurant kitchen"},
   "farmer":{"hint":"A profession","sentence":"Grows crops and cares for farm animals"},
   "driver":{"hint":"A profession","sentence":"Operates a vehicle to take people places"},
   "singer":{"hint":"A profession","sentence":"Performs songs on stage or in recordings"},
   "student":{"hint":"A person","sentence":"Learns new subjects at school or college"},
   "dog":{"hint":"It's an animal","sentence":"Friendly pet that barks"},
   "bird":{"hint":"It's an animal","sentence":"Small animal that can fly"},
   "fish":{"hint":"It's an animal","sentence":"Lives in water and swims"},
   "car":{"hint":"It's a vehicle","sentence":"Common vehicle with four wheels"},
   "train":{"hint":"It's a vehicle","sentence":"Long transport that runs on tracks"},
   "phone":{"hint":"A device","sentence":"Used to call, text, and browse"},
   "book":{"hint":"An object","sentence":"Pages you read for stories or learning"},
   "ball":{"hint":"An object","sentence":"Round toy used in many games"},
   "chair":{"hint":"An object","sentence":"You sit on it at a table"},
   "cake":{"hint":"A food","sentence":"Sweet dessert for birthdays"},
   "bread":{"hint":"A food","sentence":"Baked staple used for sandwiches"},
   "milk":{"hint":"A drink","sentence":"White drink rich in calcium"},
   "rain":{"hint":"Weather","sentence":"Water that falls from clouds"},
   "school":{"hint":"A place","sentence":"Where students go to learn"},
   "painter":{"hint":"A profession","sentence":"Creates art using colors and brushes"},
   "dancer":{"hint":"A profession","sentence":"Performs rhythm and movement on stage"},
   "jcb":{"hint":"Used to dig holes","sentence":"Yellow backhoe loader seen at construction sites"},
}
hard_words = {
   "elephant":{"hint":"Largest land animal","sentence":"Massive mammal with a trunk and tusks"},
   "aeroplane":{"hint":"Moves in the air","sentence":"Fixed-wing aircraft flying above the clouds"},
   "guitar":{"hint":"A string instrument","sentence":"Six strings, common in rock and classical music"},
   "python":{"hint":"Programming language","sentence":"Popular for automation, web apps, and data science"},
   "pilot":{"hint":"A profession","sentence":"Flies an aircraft and navigates the skies"},
   "scientist":{"hint":"A profession","sentence":"Conducts experiments to discover new knowledge"},
   "architect":{"hint":"A profession","sentence":"Designs buildings and plans their structure"},
   "journalist":{"hint":"A profession","sentence":"Reports news and writes investigative stories"},
   "engineer":{"hint":"A profession","sentence":"Builds solutions using math and technology"},
   "astronaut":{"hint":"A profession","sentence":"Trains to travel and work in space"},
   "detective":{"hint":"A profession","sentence":"Investigates cases to find facts and evidence"},
   "pharmacist":{"hint":"A profession","sentence":"Prepares and dispenses medicines safely"},
   "volcano":{"hint":"A natural feature","sentence":"Mountain that erupts with lava"},
   "microscope":{"hint":"A device","sentence":"Instrument used to see tiny objects"},
   "telescope":{"hint":"A device","sentence":"Tool used to observe distant stars"},
   "satellite":{"hint":"In space","sentence":"Object that orbits a planet"},
   "algorithm":{"hint":"In computing","sentence":"Step-by-step method to solve a problem"},
   "submarine":{"hint":"A vehicle","sentence":"Watercraft that travels under the sea"},
   "parachute":{"hint":"An object","sentence":"Helps people land safely from the air"},
   "hurricane":{"hint":"Weather","sentence":"Powerful storm with strong winds"},
   "galaxy":{"hint":"In space","sentence":"Huge group of stars and planets"},
   "compass":{"hint":"A tool","sentence":"Shows direction using north"},
   "bicycle":{"hint":"A vehicle","sentence":"Two-wheeled transport you pedal"},
   "historian":{"hint":"A profession","sentence":"Studies and writes about the past"},
   "mathematician":{"hint":"A profession","sentence":"Solves problems using numbers and logic"},
   "biologist":{"hint":"A profession","sentence":"Studies living organisms and life processes"},
}

# -------------------------
# 2) LIGHT CSS
# -------------------------
HTML("""
<style>
body{font-family:"Segoe UI",Roboto,Helvetica,Arial,sans-serif;background:#fafafa;padding-top:16px}
.game-title{font-weight:800;line-height:1.05;text-align:center;margin:0 0 28px;letter-spacing:1px}
.game-card{background:#fff;border:1px solid #e5e7eb;border-radius:24px;box-shadow:0 12px 32px rgba(0,0,0,.08);padding:28px 32px;text-align:center}
button.jp-Button{font-size:18px !important;font-weight:700 !important;border-radius:12px !important;padding:12px 20px !important}
button.jp-Button:hover{opacity:.94}
.msg{font-size:20px;padding:12px 14px;margin:14px 0 10px;border-radius:12px;border:1px solid transparent;text-align:left}
.msg.ok{background:#f0fdf4;border-color:#bbf7d0;color:#166534}
.msg.warn{background:#fffbeb;border-color:#fde68a;color:#92400e}
.msg.bad{background:#fef2f2;border-color:#fecaca;color:#991b1b}
.meta{font-size:18px;line-height:1.5;margin:6px 0 10px;text-align:left}
.footer{font-size:16px;color:#6b7280;margin-top:16px;text-align:left}
</style>
""")

# -------------------------
# 3) STATE (unchanged names)
# -------------------------
remaining_easy = list(easy_words.keys())
remaining_hard = list(hard_words.keys())
remaining_pool = None
current_bank   = None
current_mode   = None

secret = ""
custom_hint = ""
clue_sentence = ""
attempts = 0
hint_used = False

# -------------------------
# 4) UI (same widgets)
# -------------------------
def _btn(text, style="", w=BUTTON_WIDTH, h="52px"):
    return Button(description=text, button_style=style, layout=Layout(width=f"{w}px", height=h))

btn_easy   = _btn("Easy", "success")
btn_hard   = _btn("Hard", "warning")
btn_hint   = _btn("Hint", "info")
btn_submit = _btn("Submit", w=SUBMIT_WIDTH)

attempts_bar = IntProgress(value=0, min=0, max=ATTEMPTS_EASY,
                           layout=Layout(width=f"{CARD_WIDTH - BUTTON_WIDTH - 200}px", height="28px"))
lbl_attempts = Label("")
guess_box = Text(placeholder="Type your guess and press Enter",
                 layout=Layout(width=f"{CARD_WIDTH - SUBMIT_WIDTH - 40}px", height="52px"))

status_out, out = Output(), Output()
title = WHTML(value=f'<div class="game-title" style="font-size:{TITLE_SIZE}px">{TITLE_TEXT}</div>')
title.layout = Layout(margin='64px 0 24px 0')
btn_hint.disabled = True
guess_box.disabled = True

# -------------------------
# 5) HELPERS (same names)
# -------------------------
def show_message(text, kind="ok"):
    with status_out:
        clear_output()
        display(HTML(f'<div class="msg {kind}">{text}</div>'))

def show_header():
    with out:
        clear_output()
        if current_mode:
            display(HTML(f'<div class="meta"><b>Mode:</b> {current_mode.title()}</div>'))
        if clue_sentence:
            display(HTML(f'<div class="meta"><b>Guess me:</b> {clue_sentence}</div>'))

def show_word(masked_chars):
    with out:
        display(HTML(
            f'<div style="font-family:ui-monospace,Menlo,Consolas,monospace;'
            f'font-size:{WORD_SIZE}px;letter-spacing:14px;margin:24px 0 10px;">'
            f'{" ".join(masked_chars)}</div>'
        ))
        display(HTML('<div class="footer">Tip: Press Enter or click Submit.</div>'))

def update_attempts_bar():
    attempts_bar.max = ATTEMPTS_EASY if current_mode == "easy" else ATTEMPTS_HARD
    attempts_bar.value = attempts
    lbl_attempts.value = f"Attempts left: {attempts}"

# -------------------------
# 6) GAME FLOW (same names & logic)
# -------------------------
def start_round():
    """Use current mode's remaining list; no repeats; finished message when empty."""
    global secret, custom_hint, clue_sentence, attempts, hint_used
    if not current_bank:
        return
    if remaining_pool is None:
        btn_hint.disabled = True; guess_box.disabled = True
        show_message("Choose Easy or Hard to begin.", "warn")
        return
    if not remaining_pool:
        btn_hint.disabled = True; guess_box.disabled = True
        show_message(f"üéâ You finished all {current_mode.title()} words! Choose the other mode.", "ok")
        return

    secret = random.choice(remaining_pool); remaining_pool.remove(secret)
    data = current_bank[secret]
    custom_hint, clue_sentence = data["hint"], data["sentence"]
    attempts = ATTEMPTS_EASY if current_mode == "easy" else ATTEMPTS_HARD
    hint_used = False
    btn_hint.disabled = False; guess_box.disabled = False
    update_attempts_bar()

    length = len(secret)
    reveal = max(0, min(REVEAL_EASY if current_mode=="easy" else REVEAL_HARD, length))
    masked = ["_"] * length
    for i in random.sample(range(length), reveal): masked[i] = secret[i]

    show_message("New round! Good luck ‚ú®", "ok")
    show_header(); show_word(masked)

def choose_easy(_):
    """Switch to Easy (keeps used words)."""
    global current_mode, current_bank, remaining_pool
    current_mode = "easy"; current_bank = easy_words; remaining_pool = remaining_easy
    start_round()

def choose_hard(_):
    """Switch to Hard (keeps used words)."""
    global current_mode, current_bank, remaining_pool
    current_mode = "hard"; current_bank = hard_words; remaining_pool = remaining_hard
    start_round()

def show_hint(_):
    global hint_used
    if btn_hint.disabled or not current_bank: return
    if hint_used: show_message("Hint already used for this round.", "warn")
    else:
        show_message(f"Hint: {custom_hint}", "ok")
        hint_used = True
    btn_hint.disabled = True

def on_guess(_=None):
    global attempts
    g = (guess_box.value or "").strip().lower(); guess_box.value = ""
    if not g or not current_bank: return
    if g == secret:
        show_message("‚úÖ Correct! Loading next word‚Ä¶", "ok")
        guess_box.disabled = True; btn_hint.disabled = True
        time.sleep(2); start_round(); return
    attempts -= 1; update_attempts_bar()
    if attempts <= 0:
        show_message(f"‚ùå Out of tries. The word was: <b>{secret}</b>. Loading next word‚Ä¶", "bad")
        time.sleep(2); start_round()
    else:
        show_message("‚ö†Ô∏è Incorrect, try again.", "warn")

# -------------------------
# 7) WIRE + LAYOUT (same structure)
# -------------------------
btn_easy.on_click(choose_easy)
btn_hard.on_click(choose_hard)
btn_hint.on_click(show_hint)
guess_box.on_submit(on_guess)
btn_submit.on_click(on_guess)

row1 = HBox([btn_easy, btn_hard], layout=Layout(justify_content="center", gap=f"{ROW_GAP}px", width=f"{CARD_WIDTH}px"))
row2 = HBox([btn_hint, attempts_bar, lbl_attempts], layout=Layout(align_items="center", gap=f"{ROW_GAP}px", width=f"{CARD_WIDTH}px"))
row3 = HBox([guess_box, btn_submit], layout=Layout(justify_content="center", align_items="center", gap=f"{ROW_GAP}px", width=f"{CARD_WIDTH}px"))

card = VBox([row1, row2, row3, status_out, out], layout=Layout(width=f"{CARD_WIDTH}px")); card.add_class("game-card")
page = VBox([title, card], layout=Layout(align_items="center", justify_content="flex-start", width="100%", height="100vh"))
display(page)

# Initial prompt
show_message("Choose Easy or Hard to begin.", "warn")
```
